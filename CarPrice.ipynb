```python
import statsmodels.api as stats
```

```python
train_df = df.sample(frac = 0.7, random_state = 99)

test_df = df.drop(train_df.index)
```

```python
y_train = train_df.price

x_train = stats.add_constant(train_df.drop(columns = 'price'))
```

```python
model = stats.OLS(y_train, x_train)
```

```python
result = model.fit()
```

```python
print(result.summary())
```

```python
y_test = test_df.price

test_df = stats.add_constant(test_df)

x_test = test_df[x_train.columns]
```

```python
predictions = result.predict(x_test)
```

```python
y_train_new = train_df.price

x_train_new  = stats.add_constant(train_df[['enginesize', 
                                            'stroke',
                                            'horsepower',
                                            'enginetype_ohcv',
                                            'cylindernumber_five',
                                            'cylindernumber_four',
                                            'cylindernumber_six',
                                            'cylindernumber_twelve']])
                                            
```

```python
model_new  = stats.OLS(y_train_new, x_train_new)
```

```python
result_new = model_new.fit()
```

```python
print(result_new.summary())
```

```python
y_test_new = test_df.price

x_test_new = test_df[x_train_new.columns]
```

```python
new_prediction = result_new.predict(x_test_new)
```

```python
plt.scatter(new_prediction, y_test)
plt.plot([5000, 50000], [5000, 50000], ls = '--', c = 'k')
plt.xlabel('Predicted Price [$]')
plt.ylabel('Observed Price [$]')
```

```python
x_train_scale = ((x_train_new - x_train_new.mean())/x_train_new.std()).drop(columns = 'const')
```

```python
x_train_scale.mean().round(2)
```

```python
x_train_scale.std().round(2)
```

```python
x_train_scale = stats.add_constant(x_train_scale)
```

```python
model_scale = stats.OLS(y_train_new, x_train_scale)
```

```python
result_scale = model_scale.fit()
```

```python
print(result_scale.summary())
```

```python
result_scale.params.drop(index = ['const']).abs().sort_values(ascending = False).plot.bar()
```

```python
result_new.resid
```

```python
plt.scatter(result_new.fittedvalues, result_new.resid)
plt.plot([5000,50000], [0,0], ls = '--', c='k')
```

# Evaluating Linear Regression

```python
train_mse = stats.tools.eval_measures.mse(y_train_new, result_new.fittedvalues)
print('The training dataset mean square error is {}'.format(train_mse.round(1)))
```

```python
train_rmse = stats.tools.eval_measures.rmse(y_train_new, result_new.fittedvalues)
print('The training dataset root mean square error is {}'.format(train_rmse.round(1)))
```

```python
train_mae = stats.tools.eval_measures.meanabs(y_train_new, result_new.fittedvalues)
print('The training dataset mean absolute error is {}'.format(train_mae.round(1)))
```

```python
train_r2 = result_new.rsquared
print('The  training dataset coefficient of determination is {}'.format(train_r2.round(3))) 
```

```python
train_r2_adj = result_new.rsquared_adj
print('The training dataset adjusted coefficient of determination is {}'.format(train_r2_adj.round(3)))
```

The  root mean square error that our predictions are typically around \\$3000 off from the observed values and that the data explains 88.5% of the variation from the coefficient of determination. 

### Scikit-Learn

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
y = df.price

x = df[['enginesize', 
        'stroke',
        'horsepower',
        'enginetype_ohcv',
        'cylindernumber_five',
        'cylindernumber_four',
        'cylindernumber_six',
        'cylindernumber_twelve']]
                                            
```

```python
sk_x_train, sk_x_test, sk_y_train, sk_y_test = train_test_split(x, y, test_size = 0.3, random_state = 97)
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(sk_x_train, sk_y_train)
```

```python
intercept_carprice = regressor.intercept_
engsize_coeffs = regressor.coef_
ssr_carprice = np.sum((sk_y_train-regressor.predict(sk_x_train))**2)
```

```python
sk_train_predictions = regressor.predict(sk_x_train)
sk_test_predictions = regressor.predict(sk_x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
sk_train_mse = mean_squared_error(sk_y_train, sk_train_predictions)
print('The training dataset mean square error is {}'.format(sk_train_mse.round(1)))
```

```python
sk_train_rmse = np.sqrt(sk_train_mse)
print('The training dataset root mean square error is {}'.format(sk_train_rmse.round(1)))
```

```python
sk_train_mae = mean_absolute_error(sk_y_train, sk_train_predictions)
print('The training dataset mean absolute error is {}'.format(sk_train_mae.round(1)))
```

```python
sk_train_r2 = r2_score(sk_y_train, sk_train_predictions)
print('The training dataset coefficient of determination is {}'.format(sk_train_r2.round(3)))
```

```python
sk_test_r2 = r2_score(sk_y_test, sk_test_predictions)
print('The test dataset coefficient of determination is {}'.format(sk_test_r2.round(3)))
```
